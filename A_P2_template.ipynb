```python
# you will compress the file named filename, and save the compressed as filename+".lzw"
# keep the function name
def LZW_compress(fname):

    # Define the initial dictionary of ASCII values
    dictionarySize = 256
    dictionary = {chr(i): i for i in range(dictionarySize)}

    f = open(fname, "r")

    while True:
        character = f.read(1)
        print(dictionary[character])
        if character == '':
            break

    f.close()

    print("EOF")




# keep this line 
LZW_compress(filename)
```

```python
# you'll expand the file named filename.lzw, and save the decompressed as filename+".2"
# keep the function name
def LZW_expand(fname):
    print(fname)
    print(fname[:-4]+".2")
    # your code here
    
# keep this line 
LZW_expand(filename +".lzw")
```
